## Test on single simple models

### XGB Classifier

```python
def XGB_class(X_train, X_test, y_train, y_test,typ):
    
    if typ=='binary':
        clf = xgb.XGBClassifier(seed = 24, use_label_encoder =False)
    else:
        clf = xgb.XGBClassifier(seed = 24, use_label_encoder =False,objective='multi:softproba')
    
    clf.fit(X_train, y_train)
    y_predicted = clf.predict(X_test)

    print(classification_report(y_test, y_predicted))
    print("Accuracy:", accuracy_score(y_test, y_predicted))
    print("MCC =",matthews_corrcoef(y_test, y_predicted))
    
    if typ=='binary':
        print("AUC =",metrics.roc_auc_score(y_test, y_predicted))
    else:
        y_test = label_binarize(y_test, classes=np.arange(3))
        y_predicted = label_binarize(y_predicted, classes=np.arange(3))
        fpr = dict()
        tpr = dict()
        roc_auc = dict()
        for i in range(3):
            fpr[i], tpr[i], _ = roc_curve(y_test[:, i], y_predicted[:, i])
            roc_auc[i] = auc(fpr[i], tpr[i])
        print("AUC =",roc_auc)
    
    return clf.predict_proba(X_test)[:,1]
```

```python
print("binary class")
XGBc_y_predicted_2_20 = XGB_class(X_train_2_20,X_test_2_20,y_train_2_20,y_test_2_20,"binary")
XGBc_y_predicted_2_40 = XGB_class(X_train_2_40,X_test_2_40,y_train_2_40,y_test_2_40,"binary")

print("multi class")
XGBc_y_predicted_3_20 = XGB_class(X_train_3_20,X_test_3_20,y_train_3_20,y_test_3_20,"3")
XGBc_y_predicted_3_40 = XGB_class(X_train_3_40,X_test_3_40,y_train_3_40,y_test_3_40,"3")
```

### SVM

```python
def SVMc(X_train,X_test,y_train,y_test,typ):
    
    if typ=='binary':       
        clf = SVC(kernel='linear',probability=True) 
    else:
        clf = svm.SVC(kernel='poly',probability=True)

        
    clf.fit(X_train, y_train) 
    preds = clf.predict(X_test)
    sSVM_y_predicted = clf.predict_proba(X_test)[:,1]
    
    print(classification_report(y_test, preds))
    print("Accuracy:", accuracy_score(y_test, preds))
    print("MCC =",matthews_corrcoef(y_test, preds))
    
    if typ=='binary': 
        print("AUC =",metrics.roc_auc_score(y_test, preds))
    else:
        y_test = label_binarize(y_test, classes=np.arange(3))
        y_predicted = label_binarize(preds, classes=np.arange(3))
        fpr = dict()
        tpr = dict()
        roc_auc = dict()
        for i in range(3):
            fpr[i], tpr[i], _ = roc_curve(y_test[:, i], y_predicted[:, i])
            roc_auc[i] = auc(fpr[i], tpr[i])
        print("AUC =",roc_auc)
    
    return sSVM_y_predicted
```

### Random Forest classifier

```python
def RFc(X_train,X_test,y_train,y_test,typ):

    model = RandomForestClassifier().fit(X_train,y_train)
    preds = model.predict(X_test)
    sRFC_y_predicted = model.predict_proba(X_test)[:,1]
    
    print(classification_report(y_test, preds))
    print("Accuracy:", accuracy_score(y_test, preds))
    print("MCC =",matthews_corrcoef(y_test, preds))
    
    if type=='binary':
        print("AUC =",metrics.roc_auc_score(y_test, preds))
    else:
        y_test = label_binarize(y_test, classes=np.arange(3))
        y_predicted = label_binarize(preds, classes=np.arange(3))
        fpr = dict()
        tpr = dict()
        roc_auc = dict()
        for i in range(3):
            fpr[i], tpr[i], _ = roc_curve(y_test[:, i], y_predicted[:, i])
            roc_auc[i] = auc(fpr[i], tpr[i])
        print("AUC =",roc_auc)
    
    return sRFC_y_predicted
```

### Logistic Regression

```python
def LR(X_train,X_test,y_train,y_test,typ):
    if typ=='binary':
        model = LogisticRegression(random_state=0).fit(X_train,y_train)
    else:
        model = LogisticRegression(multi_class='multinomial', solver='lbfgs').fit(X_train,y_train)

    preds = model.predict(X_test)
    sLR_y_predicted = model.predict_proba(X_test)[:,1]
    
    print(classification_report(y_test, preds))
    print("Accuracy:", accuracy_score(y_test, preds))
    print("MCC =",matthews_corrcoef(y_test, preds))
    if typ=='binary':
        print("AUC =",metrics.roc_auc_score(y_test, preds))
    else:
        y_test = label_binarize(y_test, classes=np.arange(3))
        y_predicted = label_binarize(preds, classes=np.arange(3))
        fpr = dict()
        tpr = dict()
        roc_auc = dict()
        for i in range(3):
            fpr[i], tpr[i], _ = roc_curve(y_test[:, i], y_predicted[:, i])
            roc_auc[i] = auc(fpr[i], tpr[i])
        print("AUC =",roc_auc)
    return sLR_y_predicted
```

### Test on XGB classifier using RFE selected features

```python
def rfe_xgb(X_train, y_train,X_test,typ):
    min_features_to_select = 50
    
    if typ=='binary':
        clf = xgb.XGBClassifier(seed = 24, use_label_encoder =False)
    else:
        clf = xgb.XGBClassifier(seed = 24, use_label_encoder =False,objective='multi:softproba')
    
    rfecv = RFECV(estimator=clf,min_features_to_select=min_features_to_select,step=3,n_jobs=-1,scoring="r2",cv=5)
    rfecv.fit(X_train, y_train)
    
    #keep selected features + check RFE accuracy scores during running
    newX_train = X_train[X_train.columns[rfecv.support_]]
    newX_test = X_test[X_test.columns[rfecv.support_]]
    print('We kept',newX_train.shape[1],'features out of the',X_train.shape[1])
    
    return newX_train,newX_test
```

```python
print('binary')
print('Using 20% as test subset:')
newX_train,newX_test = rfe_xgb(X_train_2_20, y_train_2_20,X_test_2_20,'binary')
RFE_XGB_y_predicted_2_20 = XGB_class(newX_train,newX_test,y_train_2_20,y_test_2_20,'binary')
print('Using 40% as test subset:')
newX_train,newX_test = rfe_xgb(X_train_2_40, y_train_2_40,X_test_2_40,'binary')
RFE_XGB_y_predicted_2_40 = XGB_class(newX_train,newX_test,y_train_2_40,y_test_2_40,'binary')
```

### Plot AUC-ROC Curve

```python
def roc(y_tests,y_predicted,models_name,title):  
    if len(y_predicted)>2:  
        #there are 4 diff colors in max
        colors = ["red", "blue", "green", "orange"]
            
        for i in range(int(len(y_predicted)/2)):
            fpr, tpr, _ = metrics.roc_curve(y_tests[i],  y_predicted[i])
            #create ROC curve
            plt.plot(fpr, tpr, linestyle='-',color=colors[i])

        counter=0
        for i in range(int(len(y_predicted)/2),int(len(y_predicted))):
            fpr, tpr, _ = metrics.roc_curve(y_tests[i],  y_predicted[i])
            #create ROC curve
            plt.plot(fpr, tpr, linestyle='--',color=colors[counter])
            counter+=1
    
    else:
        for i in range(int(len(y_predicted))):
            fpr, tpr, _ = metrics.roc_curve(y_tests[i],  y_predicted[i])
            #create ROC curve
            plt.plot(fpr, tpr, linestyle='-')
    
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([-0.05, 1.0])
    plt.ylim([0.0, 1.05])
    plt.legend(models_name)
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.title(title)
    plt.show()
```

```python
y_tests = np.array([y_test_2_20,y_test_2_40])
y_preds = np.array([RFE_XGB_y_predicted_2_20,RFE_SVM_y_predicted_2_40])
models_name=np.array(['RFE_XGBc 20%','RFE_SVM 40%'])
title = 'ROC curve of stronger models using RFE'
roc(y_tests,y_preds,models_name,title)
```

### FORWARD ORDER feature selection for binary on XGBc

```python
def forward_select_2(X_train, y_train):
    
    acc_imp = 0  
    last_best_acc=0
    features_selected=[]
    best_features = []
    no_imp_counter=0
    
    while no_imp_counter<20:
        accuracies = []
        for i in range(len(X_train)):   
            curr_X_train = X_train.iloc[:,i]
            for j in range(len(features_selected)):
                curr_X_train = pd.concat([curr_X_train, features_selected[j]],axis=1)
            fs_X_train, fs_X_test, fs_y_train, fs_y_test = train_test_split(curr_X_train, y_train, test_size = 0.80, random_state = 97,stratify=y_train)

            clf = xgb.XGBClassifier(seed = 24, use_label_encoder =False)
            clf.fit(fs_X_train, fs_y_train)
            y_predicted = clf.predict(fs_X_test)
            accuracies.append(accuracy_score(fs_y_test, y_predicted))
        
        best_feature_idx=np.argmax(accuracies) 
        print('new best feature added =',X_train.iloc[:,best_feature_idx].name)
        features_selected.append(X_train.iloc[:,best_feature_idx])
        X_train.drop(X_train.columns[best_feature_idx],axis=1,inplace=True)
        
        curr_acc=accuracies[best_feature_idx]
        print('new acc =',curr_acc)
        if curr_acc>last_best_acc:
            acc_imp = curr_acc-last_best_acc
            last_best_acc = curr_acc
            best_features=features_selected.copy()
            no_imp_counter=0
            print('improvement of',acc_imp)
        else:
            no_imp_counter+=1
            print('no improvement for',no_imp_counter,'times')

    
    print('-> last best acc on training set =',last_best_acc)
    
    best_features_names = []
    for i in best_features:
        best_features_names.append(i.name)

    return best_features_names
```

```python
print('1) Forward feature selection:')
fs_best_features = forward_select_2(X_train_2_20.copy(), y_train_2_20)
newX_train = X_train_2_20.loc[:,fs_best_features]
newX_test = X_test_2_20[(newX_train.columns) & (X_test_2_20.columns)]
print('We kept',newX_train.shape[1],'features out of the',X_train_2_20.shape[1])
print('2) XGB with forward feature selection')
XGB_class(newX_train,newX_test,y_train_2_20,y_test_2_20,'binary')
```

### FORWARD ORDER feature selection for multi class on XGBc

```python
def forward_select_3(X_train, y_train):
    
    acc_imp = 0  
    last_best_acc=1000
    features_selected=[]
    best_features = []
    no_imp_counter=0
    
    while no_imp_counter<20:
        accuracies = []
        for i in range(len(X_train)):   
            curr_X_train = X_train.iloc[:,i]
            for j in range(len(features_selected)):
                curr_X_train = pd.concat([curr_X_train, features_selected[j]],axis=1)
            fs_X_train, fs_X_test, fs_y_train, fs_y_test = train_test_split(curr_X_train, y_train, test_size = 0.80, random_state = 97,stratify=y_train)

            clf = xgb.XGBClassifier(seed = 24, use_label_encoder =False,objective='multi:softproba')
            clf.fit(fs_X_train, fs_y_train)
            y_predicted = clf.predict_proba(fs_X_test)
            accuracies.append(log_loss(fs_y_test, y_predicted))
        
        best_feature_idx=np.argmin(accuracies) 
        print('new best feature added =',X_train.iloc[:,best_feature_idx].name)
        features_selected.append(X_train.iloc[:,best_feature_idx])
        X_train.drop(X_train.columns[best_feature_idx],axis=1,inplace=True)
        
        curr_acc=accuracies[best_feature_idx]
        print('new log loss =',curr_acc)
        if curr_acc<last_best_acc:
            acc_imp = last_best_acc-curr_acc
            last_best_acc = curr_acc
            best_features=features_selected.copy()
            no_imp_counter=0
            print('improvement of',acc_imp,'\n')
        else:
            no_imp_counter+=1
            print('no improvement for',no_imp_counter,'times \n')

    
    print('-> last best log loss on training set =',last_best_acc)
    
    best_features_names = []
    for i in best_features:
        best_features_names.append(i.name)

    return best_features_names
```

```python
print('1) Forward feature selection:')
fs_best_features = forward_select_3(X_train_3_20.copy(), y_train_3_20)
newX_train = X_train_3_20.loc[:,fs_best_features]
newX_test = X_test_3_20[(newX_train.columns) & (X_test_3_20.columns)]
print('We kept',newX_train.shape[1],'features out of the',X_train_3_20.shape[1])
print('2) XGB with forward feature selection')
XGB_class(newX_train,newX_test,y_train_3_20,y_test_3_20,'3')
```
